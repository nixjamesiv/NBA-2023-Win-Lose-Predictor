```python
df = {}
for year in years:
    x = 1
    link = "https://www.nba.com/stats/teams/boxscores/?Season="+year+"&SeasonType=Regular%20Season"
    driver = webdriver.Chrome(executable_path="C:\chromedriver.exe")
    driver.maximize_window()
    driver.get(link)
    driver.refresh()
    time.sleep(3)
    while x <= 50:
        soup = BeautifulSoup(driver.page_source)
        table = soup.find("table")
        df[year,x] = pd.read_html(table.prettify())[0].drop(columns=["Season", 'W/L', 'MIN', 'FGM', 'FGA', 'FG%', '3PM', '3PA', '3P%', 'FTM', 'FTA', 'FT%', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF', '+/-'])
        driver.find_element(By.CLASS_NAME, "stats-table-pagination__next").click()
        time.sleep(.5)
        x += 1
```
